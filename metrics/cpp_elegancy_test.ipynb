```python
prob_list = os.listdir(dir)
prob_list = prob_list[1:]
prob_list
```

```python
file_list = {}
for folder in prob_list:
    file_list[folder] = os.listdir(f'{dir}{folder}')

file_list
```

```python
from cpp_complexity import cal_complexity
```

```python
import numpy as np

def cal_elegance(result):
    elegance = result['number_var'] + len(result['distinct_var']) + len(result['distinct_var'])/result['number_var']
    elegance += result['number_func'] + len(result['distinct_func']) + len(result['distinct_func'])/result['number_func']
    elegance += result['LOC']
    if(result['depth']==0): print(result)
    elegance = 100*pow(elegance/100, np.log(result['depth']))
    
    return elegance
```

```python
results = {}
for folder in prob_list:
    com = []
    for file in file_list[folder]:
        if file.split('.')[-1] == 'cpp' or file.split('.')[-1] == 'c':
            com.append(cal_complexity(dir + folder + '/' + file))
            if(com[-1]['depth']==0): 
                print(file)
                print(com[-1])
            print(file, com[-1])
            results[folder] = com
        else: file_list[folder].remove(file)
#             print(file)
#             print(com)
        
print(len(results['A']))
```

```python
elegance = {}
for folder in prob_list:
    ele = []
    for result in results[folder]:
        ele.append(cal_elegance(result))
        elegance[folder] = ele
        
print(len(elegance['A']))
print(len(file_list['A']))
```
