```python
import os
```

```python
prob_list = os.listdir('./SORTED_BY_PROBLEMS')
prob_list
```

```python
file_list = {}
for folder in prob_list:
    file_list[folder] = os.listdir(f'./SORTED_BY_PROBLEMS/{folder}')

file_list
```

```python
from cpp_complexity import cal_complexity
```

```python
import numpy as np

def cal_elegancy(result):
    elegancy = result['number_var'] + len(result['distinct_var']) + len(result['distinct_var'])/result['number_var']
    elegancy += result['number_func'] + len(result['distinct_func']) + len(result['distinct_func'])/result['number_func']
    elegancy += result['LOC']
    if(result['depth']==0): print(result)
    elegancy = 100*pow(elegancy/100, np.log(result['depth']))
    
    return elegancy
```

```python
dir = './SORTED_BY_PROBLEMS/'
results = {}
for folder in prob_list:
    com = []
    for file in file_list[folder]:
        if file.split('.')[-1] == 'cpp' or file.split('.')[-1] == 'c':
            com.append(cal_complexity(dir + folder + '/' + file))
            if(com[-1]['depth']==0): 
                print(file)
                print(com[-1])
            results[folder] = com
        else: file_list[folder].remove(file)
#             print(file)
#             print(com)
        
print(len(results['A']))
```

```python
elegancy = {}
for folder in prob_list:
    ele = []
    for result in results[folder]:
        ele.append(cal_elegancy(result))
        elegancy[folder] = ele
        
print(len(elegancy['A']))
print(len(file_list['A']))
```

```python
import pandas as pd

df = {}
for folder in prob_list:
    df[folder] = pd.DataFrame({'file':file_list[folder]}, index=[i for i in range(len(file_list[folder]))])
    df[folder] = pd.concat([df[folder], pd.DataFrame({'elegancy': elegancy[folder]})], axis = 1)
```
