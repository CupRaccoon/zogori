```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
import cpp_complexity as cppcom
import data
import numpy as np
```

```python

folderList = [chr(i) for i in range(65,77)]
path = "./SORTED_BY_PROBLEMS/"
print(folderList)

def make_excel():
    fileList = {}
    for i in range(len(folderList)):
        tempFileList = os.listdir(path+folderList[i]+"/")
        cppList = [file for file in tempFileList if (file.endswith(".cpp") or file.endswith(".c"))]
        validList = []
        before_name = ['' for j in range(4)]
        for j in range(len(cppList)):
            splited_name = cppList[j].split('_')
            if (splited_name[3] != 'c.cpp' and splited_name[3] != 'C.cpp' and 
            splited_name[3] != 'c.c' and splited_name[3] != 'C.c'):
                continue

            if before_name[0] != splited_name[0] and before_name[0] != '':
                validList.append('_'.join(e for e in before_name))
            before_name = splited_name
        validList.append('_'.join(e for e in before_name))
        fileList[folderList[i]] = validList
    print(fileList)
    writer=pd.ExcelWriter('testoputput.xlsx', engine='openpyxl')

    for i in folderList:
        temp_dic = {'filename':[],'distinct_func':[], 'number_func':[],'distinct_var':[],'number_var':[],
                    'depth':[],'loc':[],'label':[]}
        for j in fileList[i]:
            print(path+i+"/"+ j)
            result = (cppcom.cal_complexity(path+i+"/"+ j))
            temp_dic['filename'].append(j)
            temp_dic['distinct_func'].append(len(result['distinct_func']))
            temp_dic['number_func'].append(result['number_func'])
            temp_dic['distinct_var'].append(len(result['distinct_var']))
            temp_dic['number_var'].append(result['number_var'])
            temp_dic['depth'].append(result['depth'])
            temp_dic['loc'].append(result['LOC'])
            temp_dic['label'].append(2)
        
            
            

        df = pd.DataFrame(temp_dic)
        df.to_excel(writer, sheet_name=i,index=False)
    writer.save()

def read_excel(filename):
    list = []
    for i in range(len(folderList)):
        list.append(pd.read_excel(filename,sheet_name=i,engine='openpyxl'))
    return list


def make_distribution(list, num, col):
    plot_max = min(list[num][col])
    x = list[num][col] == plot_max
    print(list[num][x])
    plot_max = max(list[num][col])
    x = list[num][col] == plot_max
    print(list[num][x])

    plotList = [0 for _ in range(plot_max+5)]
    for i in list[num][col]:
        plotList[i] += 1
    plt.plot(plotList)
    plt.show()

make_excel()
```

```python
labeled_data = read_excel("labeled_output.xlsx")
```

```python
labeled_list=[]
all_data = []

for i in range(len(folderList)):
    temp = []
    temp.append(labeled_data[i][labeled_data[i]['label'] != -1])
    temp.append(labeled_data[i][labeled_data[i]['label'] == 1])
    temp.append(labeled_data[i][labeled_data[i]['label'] == 2])
    temp.append(labeled_data[i][labeled_data[i]['label'] == 3])
    for j in range(4):
        temp[j] = temp[j].reset_index(drop=True)


    labeled_list.append([temp[0]])
    labeled_list[i].append(temp[1])
    labeled_list[i].append(temp[2])
    labeled_list[i].append(temp[3])
    if i == 0:
        all_data.append(temp[0])
        all_data.append(temp[1])
        all_data.append(temp[2])
        all_data.append(temp[3])
    else:
        for j in range(4):
            all_data[j] = pd.concat([all_data[j],temp[j]], ignore_index=True)
            
label_describe = [[] for _ in range(len(folderList))]
all_describe = []
for i in range(len(folderList)):
    for j in range(4):
        label_describe[i].append(labeled_list[i][j].describe())
        
for i in range(4):
    all_describe.append(all_data[i].describe())


```

```python
# print(label_describe)
# print(all_describe)

plot_list = [0 for _ in range(0,int(all_describe[0]['number_var']['max'])+1)]
bad_list = plot_list[:]
bad_count = all_describe[1]['number_var']['count']
for i in range(len(all_data[1])):
    bad_list[int(all_data[1]['number_var'][i])] += 100/bad_count

good_list = plot_list[:]
good_count = all_describe[3]['number_var']['count']
for i in range(len(all_data[3])):
    good_list[int(all_data[3]['number_var'][i])] += 100/good_count

normal_list = plot_list[:]
normal_count = all_describe[2]['number_var']['count']
for i in range(len(all_data[2])):
    normal_list[int(all_data[2]['number_var'][i])] += 100/normal_count
    
plt.plot(good_list,label="good")
plt.plot(bad_list,label="bad")
plt.plot(normal_list,label="normal")
plt.legend()
plt.show()

measure_list = ['depth']
for i in measure_list:
        plt.plot([all_describe[1][i]['mean'],all_describe[2][i]['mean'],all_describe[3][i]['mean']],label=i)
plt.legend()
plt.xticks(np.arange(0, 3, 1), labels=['bad', 'normal','good'])
plt.ylim([0, 5]) 
plt.show()

```
