```python
import os
import pandas as pd
import tensorflow as tf
import numpy as np
import keras

import matplotlib.pyplot as plt
import openpyxl
import cpp_complexity as cppcom
import data

from sklearn.model_selection import train_test_split
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K



```

```python

folderList = [chr(i) for i in range(65,77)]
path = "./SORTED_BY_PROBLEMS/"
print(folderList)

def make_excel():
    fileList = {}
    for i in range(len(folderList)):
        tempFileList = os.listdir(path+folderList[i]+"/")
        cppList = [file for file in tempFileList if (file.endswith(".cpp") or file.endswith(".c"))]
        validList = []
        before_name = ['' for j in range(4)]
        for j in range(len(cppList)):
            splited_name = cppList[j].split('_')
            if (splited_name[3] != 'c.cpp' and splited_name[3] != 'C.cpp' and 
            splited_name[3] != 'c.c' and splited_name[3] != 'C.c'):
                continue

            if before_name[0] != splited_name[0] and before_name[0] != '':
                validList.append('_'.join(e for e in before_name))
            before_name = splited_name
        validList.append('_'.join(e for e in before_name))
        fileList[folderList[i]] = validList
    writer=pd.ExcelWriter('ml_pre_data.xlsx', engine='openpyxl')

    
    for i in folderList:
        temp_dic = {'filename':[],'distinct_func':[], 'number_func':[],'distinct_var':[],'number_var':[],
                    'depth':[],'loc':[],'label':[]}
        for j in fileList[i]:
            result = (cppcom.cal_complexity(path+i+"/"+ j))
            temp_dic['filename'].append(j)
            temp_dic['distinct_func'].append(len(result['distinct_func']))
            temp_dic['number_func'].append(result['number_func'])
            temp_dic['distinct_var'].append(len(result['distinct_var']))
            temp_dic['number_var'].append(result['number_var'])
            temp_dic['depth'].append(result['depth'])
            temp_dic['loc'].append(result['LOC'])
            temp_dic['label'].append(2)
        
            
            

        df = pd.DataFrame(temp_dic)
        df.to_excel(writer, sheet_name=i,index=False)
    writer.save()

def read_excel(filename):
    list = []
    for i in range(len(folderList)):
        list.append(pd.read_excel(filename,sheet_name=i,engine='openpyxl'))
    return list


def make_distribution(list, num, col):
    plot_max = min(list[num][col])
    x = list[num][col] == plot_max
    print(list[num][x])
    plot_max = max(list[num][col])
    x = list[num][col] == plot_max
    print(list[num][x])

    plotList = [0 for _ in range(plot_max+5)]
    for i in list[num][col]:
        plotList[i] += 1
    plt.plot(plotList)
    plt.show()


```

```python
df = read_excel('ml_labeled_data.xlsx')
```

```python
input_col =['distinct_func', 'number_func','distinct_var','number_var','depth','loc']
```

```python
def make_standard_df(df,input_col_list):
    for i in range(len(df)):
        # E문제에 valid한 코드가 1개 밖에 없으므로 normalize식 nan 발생, 따라서 미리 제거
        if folderList[i] == 'E':
            continue
        if i == 0:
            normal_df = (df[i][input_col_list] - df[i][input_col_list].mean())/df[i][input_col_list].std()
            normal_df['filename'] = df[i]['filename']
            normal_df['label'] = df[i]['label']

        else:
            temp_df = (df[i][input_col_list] - df[i][input_col_list].mean())/df[i][input_col_list].std()
            temp_df['filename'] = df[i]['filename']
            temp_df['label'] = df[i]['label']
            normal_df = pd.concat([normal_df,temp_df],ignore_index=True)
    return normal_df
```

```python
def make_minmax_df(df,input_col_list):
    for i in range(len(df)):
        # E문제에 valid한 코드가 1개 밖에 없으므로 normalize식 nan 발생, 따라서 미리 제거
        if folderList[i] == 'E':
            continue
        if i == 0:
            normal_df = (df[i][input_col_list] - df[i][input_col_list].min())/(df[i][input_col_list].max()-df[i][input_col_list].min())
            normal_df['filename'] = df[i]['filename']
            normal_df['label'] = df[i]['label']

        else:
            temp_df = (df[i][input_col_list] - df[i][input_col_list].min())/(df[i][input_col_list].max()-df[i][input_col_list].min())
            temp_df['filename'] = df[i]['filename']
            temp_df['label'] = df[i]['label']
            normal_df = pd.concat([normal_df,temp_df],ignore_index=True)
    return normal_df
```

```python
normal_df = make_standard_df(df,input_col)
```

```python
print(normal_df)
```

```python
train, test = train_test_split(normal_df, test_size=0.2)
train_x = train[input_col_list].to_numpy()
train_y = tf.keras.utils.to_categorical(train['label'].to_numpy()-1)
test_x = test[input_col_list].to_numpy()
test_y = tf.keras.utils.to_categorical(test['label'].to_numpy()-1)
```

```python
K.clear_session()
model = Sequential()
model.add(layers.Dense(60,activation="relu"))
model.add(tf.keras.layers.Dropout(0.3))
model.add(layers.Dense(30,activation="relu"))
model.add(tf.keras.layers.Dropout(0.3))
model.add(layers.Dense(30,activation='relu'))
model.add(tf.keras.layers.Dropout(0.3))
model.add(layers.Dense(30,activation='relu'))
model.add(tf.keras.layers.Dropout(0.3))
model.add(layers.Dense(15,activation='relu'))
model.add(layers.Dense(3,activation='softmax'))          
#모델 컴파일
adam = keras.optimizers.Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(optimizer=adam,loss='categorical_crossentropy',metrics=["accuracy"])
```

```python
model.fit(train_x,train_y,epochs=30)
model.summary()
#모델 예측 및 평가
score = model.evaluate(test_x,test_y)
print("score:",score)
```
